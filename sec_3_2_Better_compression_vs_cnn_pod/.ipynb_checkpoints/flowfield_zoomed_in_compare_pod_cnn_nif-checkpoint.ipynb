```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

# plt.style.use('siads')
```

# compare them for 5 time snapshots
- True at 512x256
- POD 256x128 rank 1
- CNN 64x32 rank 1
- CNN 128x64 rank 1
- CNN 256x128 rank 1
- NIF rank 1

## 1. read POD model

```python
dir_name = 'RT_POD_128x256_RANK_1'
tmp=np.load('./POD/'+dir_name+'/pred.npz')
test_pod_pred_projected_on_hr = tmp['test_pod_pred_projected_on_hr']
test_pod_true_projected_on_hr = tmp['test_pod_true_projected_on_hr']
test_true_on_hr = tmp['test_true_on_hr']
```

## 2. read CNN model

### 64x32

```python
dir_name = 'RT_CNN_32x64_16,32,64_RANK_1_ACT_swish'
tmp=np.load('./CNN/'+dir_name+'/pred.npz')
test_cnn_pred_projected_on_hr_64x32 = tmp['test_pod_pred_projected_on_hr']
test_cnn_true_projected_on_hr_64x32 = tmp['test_pod_true_projected_on_hr']
test_true_on_hr = tmp['test_true_on_hr']
```

### 128x64

```python
dir_name = 'RT_CNN_64x128_16,32,64_RANK_1_ACT_swish'
tmp=np.load('./CNN/'+dir_name+'/pred.npz')
test_cnn_pred_projected_on_hr_128x64 = tmp['test_pod_pred_projected_on_hr']
test_cnn_true_projected_on_hr_128x64 = tmp['test_pod_true_projected_on_hr']
test_true_on_hr = tmp['test_true_on_hr']
```

### 256x128

```python
dir_name = 'RT_CNN_128x256_16,32,64_RANK_1_ACT_swish'
tmp=np.load('./CNN/'+dir_name+'/pred.npz')
test_cnn_pred_projected_on_hr_256x128 = tmp['test_pod_pred_projected_on_hr']
test_cnn_true_projected_on_hr_256x128 = tmp['test_pod_true_projected_on_hr']
test_true_on_hr = tmp['test_true_on_hr']
```

## 3. read NIF model

```python
%ls NIF
```

```python
dir_name = 'RT_NIF_SIREN_NSX_128_LSX_2_NST_64_LST_2_NP_1_NSENSOR_32_ACTREG_1.0/saved_model_ckpt_800'
tmp=np.load('./NIF/'+dir_name+'/pred.npz')
test_nif_pred_projected_on_hr = tmp['test_nif_pred_projected_on_hr']
test_true_on_hr = tmp['test_true_on_hr']
```

```python
test_nif_pred_projected_on_hr.shape
```

## Plot

```python
time_index = np.array([4,10,16,22,28])-1
print('time index = ', time_index)
```

```python
fig, axs = plt.subplots(6, 5, figsize=(8, 16))


# row 1: true 
for i,ii in enumerate(time_index):
    axs[0, i].imshow(test_true_on_hr[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)

# row 2: POD 256x128 -> project on 512x256
for i,ii in enumerate(time_index):
    axs[1, i].imshow(test_pod_pred_projected_on_hr[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)

# row 3: CNN 64x32 -> project on 512x256
for i,ii in enumerate(time_index):
    axs[2, i].imshow(test_cnn_pred_projected_on_hr_64x32[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
    
    
# row 3: CNN 128x64 -> project on 512x256
for i,ii in enumerate(time_index):
    axs[3, i].imshow(test_cnn_pred_projected_on_hr_128x64[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
    
# row 4: CNN 256x128 -> project on 512x256
for i,ii in enumerate(time_index):
    axs[4, i].imshow(test_cnn_pred_projected_on_hr_256x128[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
    
    
    
# row 5: NIF 512x256 direct evaluation

for i,ii in enumerate(time_index):
    axs[5, i].imshow(test_nif_pred_projected_on_hr[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
    
# axs[1, 0].imshow(Xr_TEST[-1, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
# axs[1, 0].set_title('last snap - pred')
# axs[1, 1].imshow(X_TEST[-1, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
# axs[1, 1].set_title('last snap - true')

for i1 in range(6):
    for i2 in range(5):
        axs[i1, i2].grid(False)
        axs[i1,i2].axis('off')

plt.savefig('./pngs/compare_at_5_time_stamps.png', bbox_inches='tight')
```

## plot a zoomed in

```python
fig, axs = plt.subplots(1, 6, figsize=(8, 16))


# row 1: true 
ii = time_index[-1]
axs[0].imshow(test_true_on_hr[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)

# row 2: POD 256x128 -> project on 512x256
axs[1].imshow(test_pod_pred_projected_on_hr[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)

# row 3: CNN 64x32 -> project on 512x256
axs[2].imshow(test_cnn_pred_projected_on_hr_64x32[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
        
# row 3: CNN 128x64 -> project on 512x256
axs[3].imshow(test_cnn_pred_projected_on_hr_128x64[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
    
# row 4: CNN 256x128 -> project on 512x256
axs[4].imshow(test_cnn_pred_projected_on_hr_256x128[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
        
# row 5: NIF 512x256 direct evaluation
axs[5].imshow(test_nif_pred_projected_on_hr[ii, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
    
# axs[1, 0].imshow(Xr_TEST[-1, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
# axs[1, 0].set_title('last snap - pred')
# axs[1, 1].imshow(X_TEST[-1, :, :], origin='lower', cmap='RdBu', vmin=0.9, vmax=2.3)
# axs[1, 1].set_title('last snap - true')

for i1 in range(6):
    axs[i1].grid(False)
    axs[i1].axis('off')

    axs[i1].set_ylim([128,384])
    axs[i1].set_xlim([48,100])
        
plt.savefig('./pngs/compare_at_5_time_stamps_zoomed.png', bbox_inches='tight')
```
