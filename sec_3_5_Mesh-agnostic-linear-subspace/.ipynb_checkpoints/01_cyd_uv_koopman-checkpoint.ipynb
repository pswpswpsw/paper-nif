```python
dt = 5e-4
L_char = 7e-3
U_inf = 3
```

```python
# extract temporal coefficients
pn_list = []
for i in range(0,len(index_max_list)):
    if i == 0:
        i_min = 0
    else:
        i_min = index_max_list[i-1]+1
    pn_list.append(output_arr_pn[i_min,:])
pn = np.vstack(pn_list)

# extract the A(t)
At = pn[:,:2*RANK].reshape(-1,RANK,2)
A0 = pn[:,-2:]

# perform DMD on each component
At_u = At[:,:,0]
At_v = At[:,:,1]
```

```python
def compute_dmd(X,dt,L_char,U_inf):
    X0 = X[:-1]
    X1 = X[1:]
    A=np.linalg.lstsq(X0, X1)[0].T
    D,P=np.linalg.eig(A)
    
    print('residual = ',np.linalg.lstsq(X0, X1)[1])
    
    # a-priori temporal and spatial modes
    tm = X0 @ np.linalg.inv(P).T @ np.diag(D)
    sm = P.T
    
    
    # compute continuous DMD eigenvalue
    growth_rate = np.real(np.log(D))/dt
    freq = np.imag(np.log(D))/dt/(2*np.pi)
    
    # initial amplitute
    ampl = X0[0].reshape(1,-1) @ np.linalg.inv(P).T
    ampl = np.abs(ampl.ravel())
    
    st = freq*L_char / U_inf
    return ampl, growth_rate, st, tm, sm, D
```

## visualize DMD spectrum

```python
ampl_u, growth_rate_u, st_u, tm_u, sm_u, D_u = compute_dmd(At_u,dt,L_char,U_inf)
ampl_v, growth_rate_v, st_v, tm_v, sm_v, D_v = compute_dmd(At_v,dt,L_char,U_inf)
```

```python
theta=np.linspace(0,6.28,100)
plt.figure(figsize=(4,4))
plt.plot(np.cos(theta),np.sin(theta),'b-')
plt.scatter(np.real(D_u),np.imag(D_u),s=10,c='r')
plt.xlim([-1.5,1.5])
plt.ylim([-1.5,1.5])
plt.title('DMD discrete spectrum of U')
```

## visualize DMD modes

## 1. fluctuation part of U

```python
pxT=px.T

# flunctuation terms 20, for u
df_u = []
energy = []
for i in range(RANK):
    tmp = tm_u[:,[i]] @ sm_u[[i], :]
    
    # print('i = ', i, ' norm = ', np.linalg.norm(tmp))
    energy.append(np.linalg.norm(tmp))
    df_u_i = tmp @ pxT
    # df_u_i = sm_u[[i], :] @ pxT
    df_u_i = np.real(df_u_i * std[-2])
    df_u.append(df_u_i)

gr_u = 15*np.exp(growth_rate_u)

plt.plot(energy,'k-o')
# plt.plot(gr_u)
```

```python

# levels_list = [    
#     np.linspace(-3,3,20),
#     np.linspace(-0.5,0.5,20),
#     np.linspace(-1,1,20),
#     np.linspace(-2,2,20),

# ]
levels_list = [10]*3
for i, i_mode in enumerate([8, 6,4]):
    for j in range(9):
        plt.figure(figsize=(6,4))
        plt.tricontourf(xx, yy, df_u[i_mode][j].ravel(), cmap='seismic', levels=levels_list[i])
        plt.axis('scaled')
        st_str = f'{st_u[i_mode]:0.2f}'
        plt.title('St = '+ st_str)
        
        
        ax = plt.gca()
        ax.axes.xaxis.set_visible(False)
        ax.axes.yaxis.set_visible(False)
        
        plt.colorbar()
        
        plt.savefig('./cylinder_UV/result/'+MODEL_NAME+'/pngs/dmd/u_st_'+st_str+'_time_'+str(j)+'.png')
        plt.close()
```
