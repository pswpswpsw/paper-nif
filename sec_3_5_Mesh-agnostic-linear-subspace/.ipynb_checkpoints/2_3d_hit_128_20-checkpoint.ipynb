```python
FILENAME = './jhu_hit_128_20/3dhit-txyz-128-20.npz'
# FILENAME = 'cylinder_n.npz'

data = np.load(FILENAME)['data']
mean = np.load(FILENAME)['mean']
std = np.load(FILENAME)['std']
data_raw = data 

arr=np.diff(data_raw[:,0])>0
index_max_list=np.where(arr==1)[0]
```

## 2. load models

```python
%ls ./jhu_hit_128_20/HIT_NIF_SIREN_NSNAP_20_NSX_128_OMEGA_30.0_LSX_4_NST_50_LST_1_NP_20_BS_1000
```

```python
MODEL_NAME = 'HIT_NIF_SIREN_NSNAP_20_NSX_128_OMEGA_30.0_LSX_4_NST_50_LST_1_NP_20_BS_1000'
# MODEL_NAME = 'NIF_SIREN_NP_200'
# MODEL_NAME = 'NIF_SIREN_NP_300'
# MODEL_NAME = 'NIF_SIREN_NP_400'
model_dir = './jhu_hit_128_20/'+MODEL_NAME+'/saved_model_ckpt_100'
```

```python
import tensorflow.compat.v1 as tf
```

```python
sess = tf.Session(graph=tf.Graph())
MODEL_LOADED=tf.saved_model.loader.load(sess, ["serve"], model_dir)
graph = sess.graph

# load input tensors
INPUT_T = graph.get_tensor_by_name('input_T:0')
INPUT_X = graph.get_tensor_by_name('input_X:0')
INPUT_Y = graph.get_tensor_by_name('input_Y:0')
INPUT_Z = graph.get_tensor_by_name('input_Z:0')

# ouput tensor
OUTPUT_U = graph.get_tensor_by_name('output_u:0')

#     OUTPUT_PARA_NET = graph.get_tensor_by_name('para_net:0')

# evaluate on the training data
input_arr_t = data_raw[:,0].reshape(-1,1)
input_arr_x = data_raw[:,1].reshape(-1,1)
input_arr_y = data_raw[:,2].reshape(-1,1)
input_arr_z = data_raw[:,3].reshape(-1,1)

tmp_list = []
for i in range(20):
    print('i = ', i, ' / 20')
    for j in range(128):
        print('j = ', j, ' / 128')
        tmp = []
        for k in range(2):
            index_range = [i*128**3 + j*128**2 + k*int(128**2/2), i*128**3 + j*128**2 + (k+1)*int(128**2/2)]
            input_arr_t_ = input_arr_t[index_range[0]:index_range[1]]
            input_arr_x_ = input_arr_x[index_range[0]:index_range[1]]
            input_arr_y_ = input_arr_y[index_range[0]:index_range[1]]
            input_arr_z_ = input_arr_z[index_range[0]:index_range[1]]
            output_arr_u_ = sess.run(OUTPUT_U, feed_dict={INPUT_T:input_arr_t_, 
                                                          INPUT_Y: input_arr_y_, 
                                                          INPUT_X:input_arr_x_,
                                                          INPUT_Z:input_arr_z_
                                                         })
            tmp.append(output_arr_u_)
        output_arr_u_ = np.vstack(tmp)
        
        # print(output_arr_u_.shape)
#         input_arr_t_ = input_arr_t[i*128**3+j*128**2 : i*128**3+(j+1)*128**2]
#         input_arr_x_ = input_arr_x[i*128**3+j*128**2 : i*128**3+(j+1)*128**2]
#         input_arr_y_ = input_arr_y[i*128**3+j*128**2 : i*128**3+(j+1)*128**2]
#         input_arr_z_ = input_arr_z[i*128**3+j*128**2 : i*128**3+(j+1)*128**2]
#         output_arr_u_ = sess.run(OUTPUT_U, feed_dict={INPUT_T:input_arr_t_, 
#                                                       INPUT_Y: input_arr_y_, 
#                                                       INPUT_X:input_arr_x_,
#                                                       INPUT_Z:input_arr_z_
#                                                      })
        tmp_list.append(output_arr_u_)
output_arr_u = np.vstack(tmp_list)
```

```python
## un-normalize the data
u_mag_pred = output_arr_u * std[-1] + mean[-1]

data_raw_umag = data_raw[:,-1] * std[-1] + mean[-1]
```

```python
# save the data
np.savez('./jhu_hit_128_20/'+MODEL_NAME+'/output_arr_u.npz',u_mag_pred=u_mag_pred, data_raw_umag=data_raw_umag)
# output_arr_u = np.load('./NOAA_OISST_V2/result/'+MODEL_NAME+'/output_arr_u.npz')['output_arr_u']
```
