## 1. load training data

```python
%ls NOAA_OISST_V2
```

```python
FILENAME = './NOAA_OISST_V2/sst_train_n.npz'
# FILENAME = 'cylinder_n.npz'

data=np.load(FILENAME)['data']
mean = np.load(FILENAME)['mean']
std = np.load(FILENAME)['std']
data_raw = data 

arr=np.diff(data_raw[:,0])>0
index_max_list=np.where(arr==1)[0]
```

## 2. load models

```python
%ls ./NOAA_OISST_V2/result
```

```python
MODEL_NAME = 'NIF_SIREN_NP_100'
# MODEL_NAME = 'NIF_SIREN_NP_200'
# MODEL_NAME = 'NIF_SIREN_NP_300'
# MODEL_NAME = 'NIF_SIREN_NP_400'
model_dir = './NOAA_OISST_V2/result/'+MODEL_NAME+'/saved_model_ckpt_500'
```

```python
import tensorflow.compat.v1 as tf
```

```python

sess = tf.Session(graph=tf.Graph())
MODEL_LOADED=tf.saved_model.loader.load(sess, ["serve"], model_dir)
graph = sess.graph


# load input tensors
INPUT_T = graph.get_tensor_by_name('input_T:0')
INPUT_Y = graph.get_tensor_by_name('input_Y:0')
INPUT_X = graph.get_tensor_by_name('input_X:0')

# ouput tensor
OUTPUT_U = graph.get_tensor_by_name('output_u:0')

#     OUTPUT_PARA_NET = graph.get_tensor_by_name('para_net:0')

# evaluate on the training data
input_arr_t = data_raw[:,0].reshape(-1,1)
input_arr_x = data_raw[:,1].reshape(-1,1)
input_arr_y = data_raw[:,2].reshape(-1,1)

tmp_list = []
for i in range(1642):
    print('i = ', i, ' / 1641')
    for j in range(180):
        input_arr_t_ = input_arr_t[i*360*180+j*360 : i*360*180+(j+1)*360]
        input_arr_x_ = input_arr_x[i*360*180+j*360 : i*360*180+(j+1)*360]
        input_arr_y_ = input_arr_y[i*360*180+j*360 : i*360*180+(j+1)*360]
        output_arr_u_ = sess.run(OUTPUT_U, feed_dict={INPUT_T:input_arr_t_, INPUT_Y: input_arr_y_, INPUT_X:input_arr_x_ })
        tmp_list.append(output_arr_u_)
output_arr_u = np.vstack(tmp_list)
```

```python
%ls NOAA_OISST_V2
```

```python
# save the data
# np.savez('./NOAA_OISST_V2/result/'+MODEL_NAME+'/output_arr_u.npz',output_arr_u=output_arr_u)
output_arr_u = np.load('./NOAA_OISST_V2/result/'+MODEL_NAME+'/output_arr_u.npz')['output_arr_u']
```

## visualization

```python
from netCDF4 import Dataset
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import FormatStrFormatter

f_land = Dataset("./NOAA_OISST_V2/lsmask.nc","r")
mask = f_land.variables['mask']
data_raw_ar = data_raw.reshape(1642, 180,360,4)
output_arr_u_ar = output_arr_u.reshape(1642, 180,360)
lon=data_raw_ar[0,:,:,1] * std[1] + mean[1]
lat=data_raw_ar[0,:,:,2] * std[2] + mean[2]
```

```python
levels = np.linspace(-3,37,60,endpoint=True)
levels_e = np.linspace(-3,3,60,endpoint=True)

for i in range(1642):
    fig, axs = plt.subplots(1,3,figsize=(18,4))
    # -- U
    # pred
    sst_pred = output_arr_u_ar[i,:,:]*std[-1]+mean[-1]
    sst_pred_masked = sst_pred*mask[0,:,:] + (1-mask[0,:,:])*1e6 
    im1=axs[0].contourf(lon, lat, sst_pred_masked,cmap='jet',levels = levels)
    ## colorbar
    divider = make_axes_locatable(axs[0])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im0, cax=cax,format="%.0f")
    axs[0].grid(False)
    
    # truth            
    sst_true = data_raw_ar[i,:,:,-1]*std[-1]+mean[-1]
    sst_true_masked = sst_true*mask[0,:,:] + (1-mask[0,:,:])*1e6 
    im0=axs[1].contourf(lon, lat, sst_true_masked,cmap='jet',levels = levels)
    ## colorbar
    divider = make_axes_locatable(axs[1])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im0, cax=cax,format="%.0f")
    axs[1].grid(False)
    
    # difference 
    difference = sst_pred_masked - sst_true_masked
    im2=axs[2].contourf(lon, lat, difference,cmap='seismic',levels = levels_e)
    ## colorbar
    divider = make_axes_locatable(axs[2])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im2, cax=cax,format="%.1f")
    axs[2].grid(False)
    
    axs[0].axis('scaled')
    axs[1].axis('scaled')
    axs[2].axis('scaled')
    axs[0].set_title('Pred.')
    axs[1].set_title('Truth')
    axs[2].set_title('Err.')
    plt.title('Week = '+str(int(data_raw_ar[i,0,0,0]*std[0] + mean[0])))
    plt.savefig('./NOAA_OISST_V2/result/'+MODEL_NAME+'/pngs/sst_'+str(i)+'.png',bbox_inches = 'tight')
    plt.close()
```
