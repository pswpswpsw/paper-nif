```python
FILENAME = './cylinder_UV/cylinder_uv_siren_n_eco.npz'
# FILENAME = 'cylinder_n.npz'

data=np.load(FILENAME)['data']
mean = np.load(FILENAME)['mean']
std = np.load(FILENAME)['std']
data_raw = data 

arr=np.diff(data_raw[:,0])>0
index_max_list=np.where(arr==1)[0]
```

## 2. load models

```python
# MODEL_NAME = 'NIF_MLP_RANK_30'
# MODEL_NAME = 'NIF_MLP_RANK_20'
# MODEL_NAME = 'NIF_SIREN_RANK_30'
MODEL_NAME = 'NIF_SIREN_RANK_20'

model_dir = './cylinder_UV/result/'+MODEL_NAME+'/saved_model_ckpt_95000'
%ls ./cylinder_UV/result
```

```python
import tensorflow.compat.v1 as tf
```

```python
sess = tf.Session(graph=tf.Graph())
MODEL_LOADED=tf.saved_model.loader.load(sess, ["serve"], model_dir)
graph = sess.graph

# load input tensors
INPUT_T = graph.get_tensor_by_name('input_T:0')
INPUT_Y = graph.get_tensor_by_name('input_Y:0')
INPUT_X = graph.get_tensor_by_name('input_X:0')

# ouput tensor
# OUTPUT_U = graph.get_tensor_by_name('output_u:0')
OUTPUT_PARA_NET = graph.get_tensor_by_name('para_net:0')
OUTPUT_PHI_X = graph.get_tensor_by_name('phi_x:0')

# evaluate on the training data
input_arr_t = data_raw[:,0].reshape(-1,1)
input_arr_x = data_raw[:,1].reshape(-1,1)
input_arr_y = data_raw[:,2].reshape(-1,1)

i_min = 0
i_max = index_max_list[0]+1
output_arr_pn = sess.run(OUTPUT_PARA_NET, {INPUT_T:input_arr_t})

xx = data_raw[i_min:i_max,1]*std[1]+mean[1]
yy = data_raw[i_min:i_max,2]*std[2]+mean[2]
px = sess.run(OUTPUT_PHI_X, {INPUT_Y: input_arr_y, INPUT_X:input_arr_x })[i_min:i_max]
```

```python
## dt = approximately 0.5
dt_list = []
for i in range(99):
    i_max = index_max_list[i]+1
    dt=(data_raw[i_max,0]-data_raw[i_max-1,0])*std[0]
    dt_list.append(dt)
    # print('dt = ',dt)
    
# plt.plot(dt_list)
```

```python
dt = 5e-4
L_char = 7e-3
U_inf = 3
```

```python
# extract temporal coefficients
pn_list = []
for i in range(0,len(index_max_list)):
    if i == 0:
        i_min = 0
    else:
        i_min = index_max_list[i-1]+1
    pn_list.append(output_arr_pn[i_min,:])
pn = np.vstack(pn_list)

# extract the A(t)
At = pn[:,:40].reshape(-1,20,2)
A0 = pn[:,-2:]

# perform DMD on each component
At_u = At[:,:,0]
At_v = At[:,:,1]
```

```python
def compute_dmd(X,dt,L_char,U_inf):
    X0 = X[:-1]
    X1 = X[1:]
    A=np.linalg.lstsq(X0, X1)[0].T
    D,P=np.linalg.eig(A)
    
    print('residual = ',np.linalg.lstsq(X0, X1)[1])
    
    # a-priori temporal and spatial modes
    tm = X0 @ np.linalg.inv(P).T @ np.diag(D)
    sm = P.T
    
    
    # compute continuous DMD eigenvalue
    growth_rate = np.real(np.log(D))/dt
    freq = np.imag(np.log(D))/dt/(2*np.pi)
    
    # initial amplitute
    ampl = X0[0].reshape(1,-1) @ np.linalg.inv(P).T
    ampl = np.abs(ampl.ravel())
    
    st = freq*L_char / U_inf
    return ampl, growth_rate, st, tm, sm, D
```

## visualize DMD spectrum

```python
ampl_u, growth_rate_u, st_u, tm_u, sm_u, D_u = compute_dmd(At_u,dt,L_char,U_inf)
ampl_v, growth_rate_v, st_v, tm_v, sm_v, D_v = compute_dmd(At_v,dt,L_char,U_inf)
```

```python
theta=np.linspace(0,6.28,100)
plt.figure(figsize=(4,4))
plt.plot(np.cos(theta),np.sin(theta),'b-')
plt.scatter(np.real(D_u),np.imag(D_u),s=15*np.exp(growth_rate_u),c='r')
plt.xlim([-1.1,1.1])
plt.ylim([-1.1,1.1])
plt.title('DMD discrete spectrum of U')
```

```python
theta=np.linspace(0,6.28,100)
plt.figure(figsize=(4,4))
plt.plot(np.cos(theta),np.sin(theta),'b-')
plt.scatter(np.real(D_v),np.imag(D_v),s=15*np.exp(growth_rate_v),c='r')
plt.xlim([-1.1,1.1])
plt.ylim([-1.1,1.1])
plt.title('DMD discrete spectrum of V')
```

```python
plt.figure(figsize=(4,4))
plt.scatter(st_u, ampl_u, c='k')
plt.scatter(st_u, growth_rate_u, c='r',s=10)
plt.xlim([0,3])
plt.ylim([-1,1])
plt.title('DMD spectrum of U')
plt.xlabel('St')
```

```python
plt.figure(figsize=(4,4))
plt.scatter(st_v, ampl_v, c='k')
plt.scatter(st_v, growth_rate_v, c='r',s=10)
plt.xlim([0,3])
plt.ylim([-1,1])
plt.title('DMD spectrum of V')
plt.xlabel('St')
```

## visualize DMD modes

## 1. fluctuation part of U

```python
pxT=px.T

# flunctuation terms 20, for u
df_u = []
energy = []
for i in range(20):
    tmp = tm_u[:,[i]] @ sm_u[[i], :]
    
    # print('i = ', i, ' norm = ', np.linalg.norm(tmp))
    energy.append(np.linalg.norm(tmp))
    df_u_i = tmp @ pxT
    # df_u_i = sm_u[[i], :] @ pxT
    df_u_i = np.real(df_u_i * std[-2])
    df_u.append(df_u_i)

gr_u = 15*np.exp(growth_rate_u)

plt.plot(energy,'k-o')
# plt.plot(gr_u)
```

```python

levels_list = [    
    np.linspace(-0.5,0.5,20),
    np.linspace(-1,1,20),
    np.linspace(-2,2,20),
    np.linspace(-3,3,20),
]

for i, i_mode in enumerate([9,10,12]):
    for j in range(9):
        plt.figure(figsize=(6,4))
        plt.tricontourf(xx, yy, df_u[i_mode][j].ravel(), cmap='seismic', levels=levels_list[i])
        plt.axis('scaled')
        st_str = f'{st_u[i_mode]:0.2f}'
        plt.title('St = '+ st_str)
        
        
        ax = plt.gca()
        ax.axes.xaxis.set_visible(False)
        ax.axes.yaxis.set_visible(False)
        
        plt.colorbar()
        
        plt.savefig('./cylinder_UV/result/'+MODEL_NAME+'/pngs/dmd/u_st_'+st_str+'_time_'+str(j)+'.png')
        plt.close()
```

## 2. fluctuation part of V

```python
pxT=px.T

# flunctuation terms 20, for u
df_v = []
energy = []
for i in range(20):
    tmp = tm_v[:,[i]] @ sm_v[[i], :]
    
    # print('i = ', i, ' norm = ', np.linalg.norm(tmp))
    energy.append(np.linalg.norm(tmp))
    df_v_i = tmp @ pxT
    # df_u_i = sm_u[[i], :] @ pxT
    df_v_i = np.real(df_v_i * std[-1])
    df_v.append(df_v_i)

gr_v = 15*np.exp(growth_rate_v)

plt.plot(energy,'k-o')
# plt.plot(gr_u)
```

```python

levels_list = [    
    np.linspace(-1,1,20),
    np.linspace(-1,1,20),
    np.linspace(-2,2,20),
    np.linspace(-3,3,20),
]



for i, i_mode in enumerate([9,10,12,14]):
    for j in range(9):
        plt.figure(figsize=(6,4))
        plt.tricontourf(xx, yy, df_v[i_mode][j].ravel(), cmap='seismic', levels=levels_list[i])
        plt.axis('scaled')
        st_str = f'{st_v[i_mode]:0.2f}'
        plt.title('St = '+ st_str)
        
        
        ax = plt.gca()
        ax.axes.xaxis.set_visible(False)
        ax.axes.yaxis.set_visible(False)
        
        plt.colorbar()
        
        plt.savefig('./cylinder_UV/result/'+MODEL_NAME+'/pngs/dmd/v_st_'+st_str+'_time_'+str(j)+'.png')
        plt.close()
```
