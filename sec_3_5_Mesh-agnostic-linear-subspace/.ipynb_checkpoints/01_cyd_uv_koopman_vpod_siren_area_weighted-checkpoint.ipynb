```python
def compute_dmd(X,dt,L_char,U_inf):
    X0 = X[:-1]
    X1 = X[1:]
    A=np.linalg.lstsq(X0, X1)[0].T
    D,P=np.linalg.eig(A)

    # a-posteriori evaluation
    x_list = [X0[0,:]]
    for i in range(X.shape[0]-1):
        x_list.append(A @ x_list[i])
    Xp = np.array(x_list)

    for j in range(X.shape[1]):
        plt.figure(figsize=(16,4))
        plt.plot(X[:,j],'k-o',label='true')
        plt.plot(Xp[:,j],'r--o',label='DMD')
        plt.xlabel('time',fontsize=30)
        plt.ylabel('x-'+str(j+1),fontsize=30)
        plt.legend(loc='upper right')
        plt.savefig('./cylinder_UV/'+MODEL_NAME+'/pngs/compare_latent_rep_'+str(j)+'.png',bbox_inches='tight')
    
    print('residual = ',np.linalg.lstsq(X0, X1)[1])
    
    # a-priori temporal and spatial modes
    tm = X0 @ np.linalg.inv(P).T @ np.diag(D)
    sm = P.T
    
    
    
    # compute continuous DMD eigenvalue
    growth_rate = np.real(np.log(D))/dt
    freq = np.imag(np.log(D))/dt/(2*np.pi)
    
    # initial amplitute
    ampl = X0[0].reshape(1,-1) @ np.linalg.inv(P).T
    ampl = np.abs(ampl.ravel())
    
    st = freq*L_char / U_inf
    return ampl, growth_rate, st, tm, sm, D
```

```python
ampl, growth_rate, st, tm, sm, D = compute_dmd(At,dt,L_char,U_inf)
```

## visualize DMD modes

```python
pxT=px.T

# flunctuation terms 20, for u
df_u = []
energy = []
for i in range(RANK):
    tmp = tm[:,[i]] @ sm[[i], :]
    
    # print('i = ', i, ' norm = ', np.linalg.norm(tmp))
    energy.append(np.linalg.norm(tmp))
    df_u_i = tmp @ pxT[:10,:]
    # df_u_i = sm_u[[i], :] @ pxT
    df_u_i = np.real(df_u_i * std[-3])
    df_u.append(df_u_i)

# gr_u = 15*np.exp(growth_rate_u)

plt.figure(figsize=(2,2))
plt.plot(energy,'k-o')
# plt.plot(gr_u)
```

```python
pxT=px.T

# flunctuation terms 20, for u
df_v = []
energy = []
for i in range(RANK):
    tmp = tm[:,[i]] @ sm[[i], :]
    
    # print('i = ', i, ' norm = ', np.linalg.norm(tmp))
    energy.append(np.linalg.norm(tmp))
    df_v_i = tmp @ pxT[10:,:]
    # df_u_i = sm_u[[i], :] @ pxT
    df_v_i = np.real(df_v_i * std[-2])
    df_v.append(df_v_i)

# gr_u = 15*np.exp(growth_rate_u)

plt.figure(figsize=(2,2))
plt.plot(energy,'k-o')
# plt.plot(gr_u)
```
