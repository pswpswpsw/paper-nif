```python
import sys
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import tensorflow.compat.v1 as tf
```

## 1. load training data 

```python
# NUM_DATA = 15
# NUM_DATA = 20
NUM_DATA = 24
# NUM_DATA = 29


PLOT=True # Set this False if you need to run the inference...
# then set this to True and run all to print figures

CASE_PATH = './ks_0'

FILENAME= './DATA_FINAL/ks_train_n_{}.npz'.format(NUM_DATA)
MODEL_LIST = [
              'train_{}_swish/swish/MLP_1'.format(NUM_DATA),
              'train_{}_swish/swish/MLP_2'.format(NUM_DATA),
              'train_{}_swish/swish/MLP_3'.format(NUM_DATA),
              'train_{}_swish/swish/MLP_4'.format(NUM_DATA),
              'train_{}_swish/swish/NIF_1'.format(NUM_DATA),
              'train_{}_swish/swish/NIF_2'.format(NUM_DATA),
              'train_{}_swish/swish/NIF_3'.format(NUM_DATA),
              'train_{}_swish/swish/NIF_4'.format(NUM_DATA),
            ]

data=np.load(FILENAME)['data']
mean = np.load(FILENAME)['mean']
std = np.load(FILENAME)['std']

# using only the normalized data
data_raw = data # visualize the normalized data, how it works

N_TRAIN = int(data_raw[:,0].shape[0]/25600)

print('number of training points = ',N_TRAIN)

## 2. load testing data and make it normalized


# FILENAME = './ks_0/ks_train_n_50.npz'
FILENAME = './DATA_FINAL/ks_test_n_59.npz'

data_te=np.load(FILENAME)['data']
mean_te = np.load(FILENAME)['mean']
std_te = np.load(FILENAME)['std']

# 1. unnormalized
data_te_un = data_te*std_te + mean_te

# 2. normalize with training range
data_te_n = (data_te_un - mean)/std

N_TEST = int(data_te_n[:,0].shape[0]/25600)
```

```python
if PLOT == False:
    ## 2. load models


    def rmse(predictions, targets):
        return np.sqrt(((predictions - targets) ** 2).mean())

    def compute_error_across_all_nu(MODEL):

        model_dir = CASE_PATH + '/' + MODEL + '/saved_model_ckpt_40000'

        sess = tf.Session(graph=tf.Graph())
        MODEL_LOADED=tf.saved_model.loader.load(sess, ["serve"], model_dir)
        graph = sess.graph


        # load input tensors
        INPUT_T = graph.get_tensor_by_name('input_T:0')
        INPUT_NU = graph.get_tensor_by_name('input_NU:0')
        INPUT_X = graph.get_tensor_by_name('input_X:0')

        # ouput tensor
        OUTPUT_U = graph.get_tensor_by_name('output_u:0')


        ## evaluate training error
        PRED_U_LIST = []
        TRUE_U_LIST = []
        NU_LIST = []
        for j in range(N_TRAIN):
            # print(j)
            # evaluate on the training data
            input_arr_nu = data_raw[25600*j:25600*(j+1),0].reshape(-1,1)
            # print(np.std(input_arr_nu))
            input_arr_t = data_raw[25600*j:25600*(j+1),2].reshape(-1,1)
            input_arr_x = data_raw[25600*j:25600*(j+1),1].reshape(-1,1)
            output_arr_u = sess.run(OUTPUT_U, feed_dict={INPUT_T:input_arr_t, INPUT_NU: input_arr_nu, INPUT_X:input_arr_x })
            true_arr_u = data_raw[25600*j:25600*(j+1),3].reshape(-1,1)

            PRED_U_LIST.append(output_arr_u)
            TRUE_U_LIST.append(true_arr_u)
            NU_LIST.append(input_arr_nu[0])

            # un-normalized target
            output_arr_u = output_arr_u*std[-1] + mean[-1]
            true_arr_u = true_arr_u*std[-1] + mean[-1]

            ## plot the contour
            tt = input_arr_t.reshape(256,100)*std[2] + mean[2]
            xx = input_arr_x.reshape(256,100)*std[1] + mean[1]

            # plot training data
            LEVELS = np.linspace(1.2*np.min(true_arr_u),1.2*np.max(true_arr_u),50)

        l2_err_list = []
        for i in range(len(PRED_U_LIST)):
            uu   = PRED_U_LIST[i].reshape(256,100)
            uu_t = TRUE_U_LIST[i].reshape(256,100)
            # l2_err = np.linalg.norm(uu-uu_t) / 25600
            l2_err = rmse(uu,uu_t)
            l2_err_list.append(l2_err)

        l2_err_array_tr = np.array(l2_err_list)
        mean_err_tr = np.mean(l2_err_array_tr)
        std_err_tr = np.std(l2_err_array_tr)
        nu_tr = np.array(NU_LIST)

        ## evaluate the testing error
        PRED_U_LIST = []
        TRUE_U_LIST = []
        NU_LIST = []
        for j in range(N_TEST):
            # print(j)
            # evaluate on the training data
            input_arr_nu = data_te_n[25600*j:25600*(j+1),0].reshape(-1,1)
            # print(np.std(input_arr_nu))
            input_arr_t = data_te_n[25600*j:25600*(j+1),2].reshape(-1,1)
            input_arr_x = data_te_n[25600*j:25600*(j+1),1].reshape(-1,1)
            output_arr_u = sess.run(OUTPUT_U, feed_dict={INPUT_T:input_arr_t, INPUT_NU: input_arr_nu, INPUT_X:input_arr_x })
            true_arr_u = data_te_n[25600*j:25600*(j+1),3].reshape(-1,1)        

            # un-normalized target
            output_arr_u = output_arr_u*std[-1] + mean[-1]
            true_arr_u = true_arr_u*std[-1] + mean[-1]

            PRED_U_LIST.append(output_arr_u)
            TRUE_U_LIST.append(true_arr_u)
            NU_LIST.append(input_arr_nu[0])

        l2_err_list = []
        for i in range(len(PRED_U_LIST)):
            uu   = PRED_U_LIST[i].reshape(256,100)
            uu_t = TRUE_U_LIST[i].reshape(256,100)
            l2_err = rmse(uu,uu_t)
            # l2_err = np.linalg.norm(uu-uu_t) / 25600
            l2_err_list.append(l2_err)
            print('NU =', NU_LIST[i][0]*std[0]+mean[0])

        l2_err_array_te = np.array(l2_err_list)
        mean_err_te = np.mean(l2_err_array_te)
        std_err_te = np.std(l2_err_array_te)
        nu_te = np.array(NU_LIST)

        return mean_err_tr, std_err_tr, nu_tr, l2_err_array_tr, mean_err_te, std_err_te, nu_te, l2_err_array_te

    l2_err_array_tr_list = []
    l2_err_array_te_list = []
    for MODEL in MODEL_LIST:
        print('')
        print('==================')
        print('model = ', MODEL)
        print('==================')

        mean_err_tr, std_err_tr, nu_tr, l2_err_array_tr,mean_err_te, std_err_te, nu_te, l2_err_array_te = compute_error_across_all_nu(MODEL)

        print('MEAN ERR (TRAIN) = ', mean_err_tr)
        print('STD ERR (TRAIN) = ', std_err_tr)
        print('MEAN ERR (TEST) = ', mean_err_te)
        print('STD ERR (TEST) = ', std_err_te)

        l2_err_array_tr_list.append(l2_err_array_tr)
        l2_err_array_te_list.append(l2_err_array_te)




    ## compute error

    import seaborn as sns

    # compute the mean
    mlp_swish_tr_30_mean = np.array(l2_err_array_tr_list[0:4]).mean()
    nif_swish_tr_30_mean = np.array(l2_err_array_tr_list[4:8]).mean()

    mlp_swish_te_30_mean = np.array(l2_err_array_te_list[0:4]).mean()
    nif_swish_te_30_mean = np.array(l2_err_array_te_list[4:8]).mean()

    ## scaling the nu's
    nu_tr_ = mean[0] + std[0]*nu_tr
    nu_te_ = mean_te[0] + std_te[0]*nu_te

    id_nu_tr=np.argsort(nu_tr_,axis=0)
    id_nu_te=np.argsort(nu_te_,axis=0)

    # save the file
    np.savez('./tmp_result/result_change_amount_of_data_{}.npz'.format(NUM_DATA),
             nu_tr_=nu_tr_,
             nu_te_=nu_te_,
             l2_err_array_tr_list=l2_err_array_tr_list,
            l2_err_array_te_list=l2_err_array_te_list)
```

```python
if PLOT:
    # after saving the computed error, load the file

    # NUM_DATA=24
    l2_err_array_tr_list = np.load('./tmp_result/result_change_amount_of_data_{}.npz'.format(NUM_DATA))['l2_err_array_tr_list']
    nu_tr_ = np.load('./tmp_result/result_change_amount_of_data_{}.npz'.format(NUM_DATA))['nu_tr_']
    nu_te_ = np.load('./tmp_result/result_change_amount_of_data_{}.npz'.format(NUM_DATA))['nu_te_']
    id_nu_tr=np.argsort(nu_tr_,axis=0)
    id_nu_te=np.argsort(nu_te_,axis=0)

    # plot figure 21

    ## define the plotting function
    def plot_train_test_rmse(l2_err_array_tr_list, title, fig_filename):

        fig,axs=plt.subplots(1,2,figsize=(10,4))
        axs[0].plot(nu_tr_.ravel()[id_nu_tr], np.array(l2_err_array_tr_list[0:4]).mean(axis=0)[id_nu_tr],'ro-',label='MLP')
        axs[0].plot(nu_tr_.ravel()[id_nu_tr], np.array(l2_err_array_tr_list[4:8]).mean(axis=0)[id_nu_tr],'ko-',label='NIF')
        axs[0].legend(loc='best',fontsize=15)
        axs[0].set_yscale('log')
        axs[0].set_ylabel('rmse',fontsize=30)
        axs[0].set_xlim([0.195,0.285])
        axs[0].set_xlabel(r'$\mu$',fontsize=30)
        axs[0].set_title('train',fontsize=25)

        axs[1].plot(nu_te_.ravel()[id_nu_te], np.array(l2_err_array_te_list[0:4]).mean(axis=0)[id_nu_te],'ro-',label='MLP')
        axs[1].plot(nu_te_.ravel()[id_nu_te], np.array(l2_err_array_te_list[4:8]).mean(axis=0)[id_nu_te],'ko-',label='NIF')
        axs[1].set_yscale('log')
        # axs[1].set_ylabel('rmse',fontsize=20)
        axs[1].set_xlim([0.195,0.285])
        axs[1].set_xlabel(r'$\mu$',fontsize=30)
        axs[1].set_title('test',fontsize=25)

        axs[0].tick_params(axis='both', labelsize=25)
        axs[1].tick_params(axis='both', labelsize=25)

        fig.tight_layout()
        fig.suptitle(title,fontsize=30,x=0.25,y=1.05)

        plt.savefig('./paper-pngs/{}.png'.format(fig_filename), bbox_inches='tight')

    if NUM_DATA == 15:
        plot_train_test_rmse(l2_err_array_tr_list, '(1) train data: {} simulations'.format(NUM_DATA), 'ks_0_change_data_{}'.format(NUM_DATA))
    elif NUM_DATA == 20:
        plot_train_test_rmse(l2_err_array_tr_list, '(2) train data: {} simulations'.format(NUM_DATA), 'ks_0_change_data_{}'.format(NUM_DATA))
    elif NUM_DATA == 24:
        plot_train_test_rmse(l2_err_array_tr_list, '(3) train data: {} simulations'.format(NUM_DATA), 'ks_0_change_data_{}'.format(NUM_DATA))
    elif NUM_DATA == 29:
        plot_train_test_rmse(l2_err_array_tr_list, '(4) train data: {} simulations'.format(NUM_DATA), 'ks_0_change_data_{}'.format(NUM_DATA))



    ## plot figure 9 (left)

    dp = [15,20,24,29]
    te_err_mlp = []
    te_err_nif = []

    for i in dp:
        tmp=np.load('./tmp_result/ks_0_change_data_'+str(i)+'.npz')
        te_err_mlp.append(tmp['mlp_swish_loss'])
        te_err_nif.append(tmp['nif_swish_loss'])

    plt.figure(figsize=(4,4))
    plt.plot(dp, te_err_mlp,'ro-',label='MLP')
    plt.plot(dp, te_err_nif,'bo-',label='NIF')
    plt.xlim([13,30])
    plt.legend(loc='best',fontsize=12)

    # plt.yscale('log')
    plt.xlabel('number of training points in parameter space',fontsize=15)
    plt.ylabel('testing error',fontsize=15)
    plt.savefig('./paper-pngs/ks_0_change_data.png',bbox_inches='tight')
```
