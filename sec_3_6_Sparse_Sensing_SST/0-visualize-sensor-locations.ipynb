```python
from netCDF4 import Dataset
import numpy as np
from matplotlib import pyplot as plt
import tensorflow.compat.v1 as tf
import sys

# plt.style.use('siads')
%matplotlib inline
```

## read the original data

```python
rootgrp = Dataset("./DATA/sst.wkmean.1990-present.nc","r")
f_land = Dataset("./DATA/lsmask.nc","r")

x = rootgrp.variables['lon'][:]
y = rootgrp.variables['lat'][:]
u = rootgrp.variables['sst'][:]
t = np.arange(u.shape[0])

# mask out these land
mask = f_land.variables['mask'][0,:,:]
xx,yy = np.meshgrid(x,y)

# mask out x,y,u
xx_s = xx[mask==1]
yy_s = yy[mask==1]

xx_l = xx[mask!=1]
yy_l = yy[mask!=1]

SNAP=832
# convert np.ma.ndarray to np.ndarray
u_arr = u[:SNAP,mask==1].filled(0)
```

```python
def visualize_sst_optimal_sensor(N_SENSOR, filename, u):
    
    sensor_path = './QDEIM/' + str(N_SENSOR) + '_sensor_location.npz'
    i_sensor = np.load(sensor_path)['i_sensor']

    size_land = 1
    size_sea = 1
    
    fig,axs=plt.subplots(1,2,figsize=(14,4))
    levels = np.linspace(-3,37,100)

    ## true sensor   
    # -land shadow
    im = axs[0].scatter(xx_l,yy_l,s=size_land,c=5*np.ones(yy_l.shape),vmin=-3,vmax=37,cmap='Greys',marker='o')
    im = axs[0].scatter(xx_s[i_sensor],yy_s[i_sensor],s=5*size_sea,c=u[i_sensor],vmin=-3,vmax=37,cmap='jet',marker='o')
    plt.colorbar(im,ax=axs[0])
    axs[0].set_title('Sensors',size=15)

    ## true field
    # -land shadow
    im = axs[1].scatter(xx_l,yy_l,s=size_land,c=5*np.ones(yy_l.shape),vmin=-3,vmax=37,cmap='Greys',marker='o')
    im = axs[1].scatter(xx_s,yy_s,s=size_sea,c=u,vmin=-3,vmax=37,cmap='jet',marker='o')
    plt.colorbar(im,ax=axs[1])
    axs[1].set_title('SST',size=15)

    ## shadow
#     im = axs[2].scatter(xx_l,yy_l,s=size,c=10*np.ones(yy_l.shape),vmin=-3,vmax=37,cmap='Greys')
#     plt.colorbar(im,ax=axs[2])
#     axs[2].set_title('Land',size=15)

    for i in range(2):
        axs[i].set_xlim([0.0,360])
        axs[i].set_ylim([-90,90])
        axs[i].set_xticks([])
        axs[i].set_yticks([])
    
    # save
    plt.savefig('png-vis/'+filename+'.png',bbox_inches='tight')
```

## visualize the sensor locations at for example, 50 sensors

```python
N_SENSOR = 50
visualize_sst_optimal_sensor(N_SENSOR, '50_sensor_first', u_arr[0])
```
