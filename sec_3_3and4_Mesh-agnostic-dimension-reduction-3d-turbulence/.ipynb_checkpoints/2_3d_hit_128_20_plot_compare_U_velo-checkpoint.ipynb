```python
data_pred = h5py.File(PATH+'/test_pred.h5','r')
data_pred.keys()
```

```python
data_true = h5py.File(PATH+'/test.h5','r')
data_true.keys()
```

```python
import matplotlib.ticker as ticker

y = data_pred['ycoor']
z = data_pred['zcoor']
yy,zz=np.meshgrid(y,z)

ix_list = [0,32,64]
ic = 0

fig,axs=plt.subplots(len(ix_list),3,figsize=(16,4*len(ix_list)))
# fig,axs=plt.subplots(len(ix_list),3,figsize=(16,16))

levels = np.linspace(-1.4,1.4,51,endpoint=True)
# size =0.5

for i in range(len(ix_list)):
    ix = ix_list[i]
    
    true = data_true['Velocity_0001'][ix,:,:,ic]
    pred = data_pred['Velocity_0001'][ix,:,:,ic]
    err = true - pred

    im = axs[i,0].contourf(yy,zz,true,levels=levels,cmap='bwr')
    # plt.colorbar(im,ax=axs[i,0 ])

    ## pred
    im = axs[i,1].contourf(yy,zz,pred,levels=levels,cmap='bwr')
    # plt.colorbar(im,ax=axs[i, 1])
    
#     if i == 0:
#         axs[i, 0].set_title('True - U',size=25)
#         axs[i, 1].set_title('Pred - U',size=25)
#         axs[i, 2].set_title('Err',size=25)

    ## loss
    im = axs[i, 2].contourf(yy,zz,err*5,levels=levels,cmap='bwr')
    
    # cbar = plt.colorbar(im,ax=axs[i, 2],format=ticker.FuncFormatter(lambda x,y: '{0:.2f}'.format(x)))
    # cbar.ax.tick_params(labelsize=20)
    
    for k in range(3):
        axs[i,k].set_xticks([])
        axs[i,k].set_yticks([])
        axs[i,k].set_xlim([yy.min(),yy.max()])
        axs[i,k].set_ylim([zz.min(),zz.max()])

plt.tight_layout()
# plt.savefig('./pngs/compare_U_pred_true-new.png',bbox_inches='tight')
# plt.savefig('./pngs/compare_U_pred_true-new-colorbar.png',bbox_inches='tight')
```
